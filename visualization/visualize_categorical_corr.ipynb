# Functions

```python
def cramers_v(x, y):
    """
    For finding correlations between categorical data.
    Cramers V is symmetric, which decreases information shown.
    0 is no association, 1 is full association
    
    :source: https://github.com/shakedzy/dython/blob/master/dython/nominal.py
    """

    confusion_matrix = pd.crosstab(x, y)
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
    rcorr = r-((r-1)**2)/(n-1)
    kcorr = k-((k-1)**2)/(n-1)
    return np.sqrt(phi2corr/min((kcorr-1), (rcorr-1)))

def conditional_entropy(x, y):
    """
    Calculates the conditional entropy of x given y: S(x|y)
    Wikipedia: https://en.wikipedia.org/wiki/Conditional_entropy
    :param x: list / NumPy ndarray / Pandas Series
        A sequence of measurements
    :param y: list / NumPy ndarray / Pandas Series
        A sequence of measurements
    :return: float
    
    :source: https://github.com/shakedzy/dython/blob/master/dython/nominal.py
    """
    
    # entropy of x given y
    y_counter = Counter(y)
    xy_counter = Counter(list(zip(x,y)))
    total_occurrences = sum(y_counter.values())
    entropy = 0.0
    for xy in xy_counter.keys():
        p_xy = xy_counter[xy] / total_occurrences
        p_y = y_counter[xy[1]] / total_occurrences
        entropy += p_xy * math.log(p_y/p_xy)
    return entropy

def theils_u(x, y):
    """
    Aka uncertainty coefficient.
    Given value of x, calculate the possible states of y and their occurence.
    Assymetrical
    
    :source: https://github.com/shakedzy/dython/blob/master/dython/nominal.py
    """
    s_xy = conditional_entropy(x, y)
    x_counter = Counter(x)
    total_occurrences = sum(x_counter.values())
    p_x = list(map(lambda n: n/total_occurrences, x_counter.values()))
    s_x = ss.entropy(p_x)
    if s_x == 0:
        return 1
    else:
        return (s_x - s_xy) / s_x
    
```

```python
def check_with_theils(cols):
    """
    Check if the low cramers v categories also have low theils

    :param cols: list of columns to check
    :type  cols: list
    """

    for col in cols:
        calc_theils = theils_u(train_X[col], train_y['status_group'])
        print(col, '\n', calc_theils)
        calc_theils = theils_u(train_y['status_group'], train_X[col])
        print(calc_theils)


def get_cat_corr_with_label(dfx, dfy):
    """
    Create a heat map for categorical correlations using Cramers V.

    :param dfx: dataframe of inputs
    :type  dfx: pandas.core.frame.DataFrame
    :param dfy: dataframe of labels
    :type  dfy: pandas.core.frame.DataFrame
    """

    cat_cols = dfx.select_dtypes(
        include=['category', 'object']
    )

    #corrs = []
    corrs = {}
    for col in cat_cols.columns:
        try:
            corr = cramers_v(dfx[col].dropna(), dfy['status_group'])
        except ZeroDivisionError:
            corrs[col] = 0
        corrs[col] = corr

        print(col, corr)
    corrs = pd.DataFrame(corrs, index=[1])
    plt.figure(figsize=(20, 2))
    sns.heatmap(corrs, annot=True)


def theils_heat_map(dfx):
    """
    Create a heat map for categorical correlations using theils u.

    :param dfx: dataframe of inputs
    :type  dfx: pandas.core.frame.DataFrame
    """

    cat_cols = dfx.select_dtypes(
        include=['category', 'object']
    ).sort_index(axis=1)

    corrs = {}
    for col in cat_cols.columns:
        col_corr = []
        for col2 in cat_cols.columns:
            try:
                corr = theils_u(dfx[col], dfx[col2])
            except ZeroDivisionError:
                corr = 0

            col_corr.append(corr)
        corrs[col] = col_corr

    corrs = pd.DataFrame(corrs, index=sorted(
        cat_cols.columns)).sort_index(axis=1)
    plt.figure(figsize=(20, 20))
    sns.heatmap(corrs, annot=True)


def check_corr_with_cramers(check_cases):
    """
    Check to see if the correlations are also high for cramer V.

    :param check_cases: dictionary of cases to check, where keys are filler names 
                        and values are a list of categories to perform cramer V on
    :type  check_cases: dict
    """

    for key, col_list in check_cases.items():
        print(key)
        for i, col in enumerate(col_list):
            try:
                cramer = cramers_v(train_X[col], train_X[col_list[i + 1]])
                print(cramer)
            except IndexError:
                pass
```

```python
def convert_num_labels_to_cat (df, categories):
    """
    Some columns have integer values, but do not denote any order, 
    so we need to convert them into categories.
    
    :param df: dataframe
    :type  df: pandas.core.frame.DataFrame
    :param categories: list of column names to convert
    :type  categories: list
    """
    
    for cat in categories:
        df[cat] = df[cat].astype('category')
    
    return df
```
