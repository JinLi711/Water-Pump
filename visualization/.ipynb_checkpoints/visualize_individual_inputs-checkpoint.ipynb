# Functions

```python
def bargraph(col, data=train_X):
    """
    Plot a bar graph

    :param data: Pandas dataframe
    :type  data: pandas.core.frame.DataFrame
    :param col: Column name
    :type  col: str
    :param title: Title to label the graph
    :type  title: str
    """

    count_types = data[col].value_counts()

    plt.figure(
        figsize=(24, 10)
    )
    sns.barplot(
        count_types.index,
        count_types.values,
        alpha=0.7
    )
    plt.title(
        col,
        fontsize=20
    )
    plt.ylabel(
        'Number of Occurrences',
        fontsize=14
    )
    plt.xlabel(
        'Types',
        fontsize=14
    )
    
    plt.show()
```

```python
def head_value_counts(col):
    """
    Return the first 5 rows of the unique value counts of a certain column
    
    :param col: column name
    :type col: str
    :returns: a series where the index are the first five unique values and the values are the frequencies
    :rtype:   pandas.core.series.Series
    """
    
    head = train_X[col].value_counts()
    print (len (head))
    return head.head()
```

```python
def histogram (label):
    """
    Create a histogram given a column from a dataframe.
    
    :param label: column name
    :type  label: str
    """
    
    plt.figure(
        figsize=(15,10)
    )
    sns.distplot(train_X[label], kde=False, )
    plt.title('Frequency Histogram')
    plt.show()
    plt.clf()
```

```python
plt.figure(
    figsize=(15,10)
)
sns.distplot(amount_tsh)
plt.title('Density Histogram')
plt.show()
plt.clf()
```

```python
# Huge outliers, so we will use z score threshold of 3 to remove it

z = np.abs(stats.zscore(amount_tsh))

plt.figure(
    figsize=(15,10)
)
sns.distplot(amount_tsh[z < 3], kde=False,)
plt.title('Frequency Histogram With Z Score Less Than 3')
plt.show()
plt.clf()
```

```python
# z score of 2
plt.figure(
    figsize=(15,10)
)
sns.distplot(amount_tsh[z < 2])
plt.title('Frequency Histogram With Z Score Less Than 2')
plt.show()
plt.clf()
```
